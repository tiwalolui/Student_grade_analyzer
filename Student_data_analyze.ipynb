```python
import pandas as pd
from google.colab import files
uploaded = files.upload()

df=pd.read_csv("student_data.csv")
df[["G1","G2","G3"]].head()



max_score=df["G3"].max()
min_score=df["G3"].min()
sex=df.groupby("sex")["G3"].mean()
age_count=df["age"].value_counts
Medu=df.groupby("Medu")["G3"].mean()
Fedu=df.groupby("Fedu")["G3"].mean()
travel_time=df.groupby("traveltime")["G3"].mean()
study_time=df.groupby("studytime")["G3"].mean()
support=df.groupby(["schoolsup", "famsup"])["G3"].mean()
activities=df.groupby("activities")["G3"].mean()
internet=df.groupby("internet")["G3"].mean()
freetime=df.groupby("freetime")["G3"].mean()
absences=df.groupby("absences")["G3"].mean()

print(max_score,min_score,sex,age_count,Medu,Fedu,travel_time,study_time,support, activities,internet,freetime,absences)
```
